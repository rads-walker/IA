# Defina suposição de Markov.

**R:** A suposição é que se pode se definir o estado atual baseado num numero finito de estados anteriores.

# Defina modelos de transição.

**R:** Temos então um modelo que tem regras baseadas em probabilidade para transições entre estados, sempre a definição do proximo estado é tomado baseado em possiveis estados atuais.

# Defina cadeia de Markov.

**R:** E uma sequencia de resultados obtidos no qual a distribuição de cada resultado segue a suposição de Markov.


Temos que Markov definiu uma ideia de que o resultado que temos como "Presente" são condicionalmente independe dos resultados obtivos no "Passado" e do "Futuro". Todas as definições de proximos estados são obtidas baseadas nos possiveis estados atuais.
